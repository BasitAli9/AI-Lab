# 1.	Write program to perform Depth First Search. This code should print nodes traversal order and adjacent values for every nodes.

```python
def dfs(node, graph, visited, component):
    visited[node] = True
    component.append(node)

    for neighbour in graph[node]:
        if not visited[neighbour]:
            dfs(neighbour, graph, visited, component)


if __name__ == "__main__":

    # Graph of nodes
    graph = {
        0: [1,2,3],
        1: [0],
        2: [0, 3, 4],
        3: [0,2],
        4: [2]
    }

    node = 0  # Starting node
    visited = [False] * len(graph)  # Make all nodes False initially
    component = []

    dfs(node, graph, visited, component)  # Traverse each node of graph

    print(f"Following is the Depth-first search: {component}")
```

# 2 Breath First Search

```python
from collections import deque

def bfs(start_node, graph, visited, component):
    # Initialize the queue with the starting node
    queue = deque([start_node])
    visited[start_node] = True # Mark the start node as visited

    while queue:
        # Pop the first node added to the queue (FIFO)
        node = queue.popleft()
        component.append(node) # Store answer

        # Traverse each adjacent node of the current node
        for child in graph[node]:
            if not visited[child]:
                visited[child] = True # Mark as visited before adding to queue
                queue.append(child)

if __name__ == "__main__":
    # Graph of nodes
    graph = {
        0: [2],
        1: [2, 3],
        2: [0, 1, 4],
        3: [1, 4],
        4: [2, 3]
    }

    start_node = 0
    visited = [False] * len(graph)
    component = []

    bfs(start_node, graph, visited, component)
    print(f"Following is the Breadth-first search: {component}")
```
