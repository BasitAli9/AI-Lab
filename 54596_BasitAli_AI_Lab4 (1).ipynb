#Dictionaries of Python

```python
# Defining the dictionary
dict1 = {
    'Name': 'Basit',
    'Age': 21,
    'Class': 'Nine'
}

print("Dictionary No 1: ", dict1)
print("Type of dict1 is :", type(dict1))

# Accessing Items using square brackets
print("dict1['Name']: ", dict1['Name'])
print("dict1['Age']: ", dict1['Age'])

# Accessing items using the .get() method
x = dict1.get('Age')
print("Getting item through get method", x)

# Getting all keys using the .keys() method
y = dict1.keys()
print("Getting keys through get method", y)
```

# Update Dictionary

```python
# 1. Initialize the dictionary
student = {
    'Name': 'Basit',
    'Age': 21,
    'Class': 'Nine'
}
print(f"Original: {student}")

# 2. Update via Direct Assignment
# Best for changing a single known key
student['Age'] = 18
print(f"Direct Assignment: {student}")

# 3. Update via .update() method
# Best for updating multiple values at once or merging dictionaries
student.update({"Age": 17, "City": "lahore"})
print(f"Using .update(): {student}")

# 4. Update via Merge Operator (Python 3.9+)
# This creates a new dictionary without modifying the original 'student'
new_data = {"Age": 17, "Grade": "A"}
student = student | new_data
print(f"Using Merge (|): {student}")
```

# Removing Items

```python
# Initializing the dictionary
dict1 = {
   'Name': 'Basit',
    'Age': 21,
    'Class': 'Nine'
}

# Displaying the original dictionary
print("Dictionary No 1: ", dict1)

# Removing the item with the key 'Class' using pop()
dict1.pop('Class')

# Displaying the dictionary after removal
print("Dictionary after pop (class): ", dict1)
```

# Removing Dictionary using Popitem

```python
dict1 = {
    'Name': 'Basit',
    'Age': 21,
    'Class': 'Nine'
}

# Removes the last item and stores it
removed_item = dict1.popitem()

print(f"Removed: {removed_item}")
print(f"Updated Dictionary: {dict1}")
```

# Deleting item Through Del

```python
dict1 = {
    'Name': 'Ali',
    'Class': 'Eight',
    'Age': 18
}

print("Dictionary No 1: ", dict1)

# Deleting the item with the key 'Class'
del dict1['Class']

print("Delete item using del : ", dict1)
```

## Task # 01

Creating  Numpy array, spliting  into 3 arrays then calculating sum of all arrays numbers.

```python
import numpy as np

# Create an array
arr = np.array([10, 27, 57, 35, 56, 60, 73, 78, 91])

# Split into 3 arrays
split_arr = np.array_split(arr, 3)

# Calculate and print sums
for i, sub_arr in enumerate(split_arr, 1):
    print(f"Array {i}: {sub_arr}, Sum: {np.sum(sub_arr)}")
```

# Task # 02

Creating a NumPy array to extract all numbers greater than 5 and less than 15.

```python
import numpy as np

arr = np.array([1, 5, 7, 9, 10, 13, 16 ,20])

# Extract numbers > 5 and < 15
filtered_arr = arr[(arr > 5) & (arr < 15)]

print("Filtered Array:", filtered_arr)
```

# Task # 04

Creating a dictionary named “students”. Store 5 different students record into dictionary. For key used sapId of the student. Student record should contain: sapid, name, age and Semester (Six). Update the Semester (Six) to Semester (Seven).

```python

students = {
    1: {"sapid": 54596, "name": "Basit", "age": 21, "semester": "Six"},
    2: {"sapid": 54131, "name": "Shahroz", "age": 22, "semester": "Six"},
    3: {"sapid": 54915, "name": "Sarmad", "age": 20, "semester": "Six"},
    4: {"sapid": 54506, "name": "Ali", "age": 20, "semester": "Six"},
    5: {"sapid": 54323, "name": "Iqbal", "age": 19, "semester": "Six"}
}

# Update Semester from "Six" to "Seven" for all students
for sap_id in students:
    if students[sap_id]["semester"] == "Six":
        students[sap_id]["semester"] = "Seven"

# Display one record to verify
print("Updated Record for ID 3:", students[3])
```
